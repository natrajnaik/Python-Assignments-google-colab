```python
# fill null values
# number --> mean
# object --> mode

for c in d.columns:
  if d[c].dtype == object:
    filler = d[c].mode()[0] 
    d[c].fillna(filler, inplace=True) 
  else: # int, float
    filler = d[c].mean()
    d[c].fillna(filler, inplace=True)

```
