```python
import pandas as pd
import numpy as np
```

```python
d={'a':[1,2,3,4,5],'b':[5,4,3,5,6],'c':[3,5,6,6,7]}
```

```python
ind=['c','d','e','f','g']
df=pd.DataFrame(d,index=ind,columns=['b','a','c'])
```

```python
df=df.rename(columns={'b':'p','a':'q','c':'r'})
```

```python
df=pd.DataFrame(d,ind)
```

```python
# Use of Loc and Iloc 
# Loc.. means location, can find location of any entities present in a datframe
# Updating a vlaue to a given entity

df.loc['e','a']=10
```
