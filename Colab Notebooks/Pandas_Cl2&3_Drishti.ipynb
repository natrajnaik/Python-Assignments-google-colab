```python
type(df)
```

```python
plt.scatter(x=df['PetalLengthCm'],y=df['PetalWidthCm'])
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('PetalLengthCm vs PetalWidthCm')
```

```python
#Addition of Hue
#Seaborn has hue property, whereas matplotlib doesn't
sns.scatterplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],hue=df['Species'])
```

```python
# Countplot
sns.countplot(x=df['Species'],data=df)
```

```python
#Pairplot
sns.pairplot(df,hue='Species')
```

```python
df.head()
```

```python
# Species column has 4 unique entities and when we use pair plot for species, we should expect 4*4 = 16 plots
# However, there is an extra column 'id' present in the df which has nothing to do with pair plot, better drop that from df
sns.pairplot(df.drop('Id',axis=1),hue='Species')
```

```python
# Boxplot
sns.boxplot(x='Species',y=df['PetalLengthCm'],data=df)
```

```python
sns.boxplot(x='Species',y=df['PetalWidthCm'],data=df)
```

```python
# Correlation
df.corr()
```

```python
# Heatmap
sns.heatmap(df.drop('Id',axis=1).corr(),annot=True)
```

```python
# Seaborn distplot
sns.distplot(df['SepalLengthCm'],color='orange')
```

```python
sns.distplot(df['SepalLengthCm'],color='red',kde=False,bins=3) #Analysing only histogramplot
```

```python
sns.distplot(df['SepalLengthCm'],color='green',hist=False,vertical=True)
```

```python
# Seaborn Jointplot..> we can plot histogram over scatter plot
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=df,color='olive',kind='reg')  # If you provide either df['string'] or 'string' it should return the plot
```

```python
# Boxplot
sns.boxplot(x=df['Species'],y=df['PetalWidthCm'],data=df,linewidth=4)
```

```python
# Lineplot
sns.lineplot(x=df['PetalLengthCm'],y=df['PetalWidthCm'],data=df,hue='Species',style='Species',markers=True)
```

```python
# Barplot
sns.barplot(x=df['Species'],y=df['PetalWidthCm'],data=df,hue='Species')
```

```python
#Palette Attribute
sns.barplot(x=df['Species'],y=df['PetalWidthCm'],data=df,palette='Blues_d')
```
