```python
with open("/content/drive/MyDrive/Colab Notebooks/covid.csv",'r') as f:
  reader=csv.reader(f)
  cols=next(iter(reader))
  rows=[]
  for row in reader:
    rows.append(row)
  
```

```python
# Writing headers to a csv file
with open("/content/drive/MyDrive/Colab Notebooks/covid.csv",'w') as f1:
  writer=csv.writer(f1)

  # writing a single header to csv file
  writer.writerow(['State', 'Total Deaths', 'Total Discharged'])

  # writing data following this header
  for k,v in d.items():
    writer.writerow([k,*v])


```
